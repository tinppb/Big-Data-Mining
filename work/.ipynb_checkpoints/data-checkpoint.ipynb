# Tạo dữ liệu bài 1

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 1. Cấu hình quy mô dữ liệu
num_rows = 1000000  # 1 triệu dòng
num_stores = 50     # 50 cửa hàng
num_products = 500  # 500 mã sản phẩm khác nhau

# 2. Danh mục sản phẩm mẫu
categories = ['Electronics', 'Grocery', 'Fashion', 'Home & Garden', 'Beauty']

# 3. Tạo dữ liệu danh mục cho sản phẩm
product_cats = {i: np.random.choice(categories) for i in range(1, num_products + 1)}

# 4. Sinh dữ liệu ngẫu nhiên
data = {
    'transaction_id': np.arange(1, num_rows + 1),
    'store_id': np.random.randint(1, num_stores + 1, size=num_rows),
    'product_id': np.random.randint(1, num_products + 1, size=num_rows),
    'quantity': np.random.randint(1, 10, size=num_rows),
}

# Tạo DataFrame
df = pd.DataFrame(data)

# 5. Map category và gán giá ngẫu nhiên dựa trên sản phẩm
df['category'] = df['product_id'].map(product_cats)
df['price'] = np.random.uniform(5.0, 500.0, size=num_rows).round(2)

# 6. Tạo transaction_date trong khoảng 1 năm trở lại đây
start_date = datetime(2025, 1, 1)
df['transaction_date'] = [start_date + timedelta(seconds=np.random.randint(0, 31536000)) 
                          for _ in range(num_rows)]

# Sắp xếp theo thời gian cho logic
df = df.sort_values('transaction_date').reset_index(drop=True)

# 7. Xuất ra file CSV (hoặc Parquet để tối ưu Big Data)
df.to_csv('../data/retail_transactions_large.csv', index=False)
print(df.head())
print(f"\nTổng số dòng: {len(df)}")
```

# Tạo dữ liệu bài 2

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os

# 1. Cấu hình
num_rows = 1000000
pages = ['home', 'product_list', 'product_detail', 'cart', 'search', 'checkout', 'payment']
actions = ['view', 'click', 'scroll', 'add_to_cart']
devices = ['mobile', 'desktop', 'tablet']
countries = ['Vietnam', 'USA', 'Thailand', 'Singapore', 'Japan', 'Korea']

# 2. Sinh dữ liệu ngẫu nhiên
data = {
    'user_id': [f"U{np.random.randint(1, 50000):05d}" for _ in range(num_rows)],
    'page': np.random.choice(pages, size=num_rows),
    'action': np.random.choice(actions, size=num_rows),
    'device': np.random.choice(devices, size=num_rows),
    'country': np.random.choice(countries, size=num_rows)
}

df_pandas = pd.DataFrame(data)

# 3. Tạo timestamp ngẫu nhiên trong 90 ngày gần đây
start_date = datetime(2026, 1, 1)
df_pandas['timestamp'] = [start_date + timedelta(seconds=np.random.randint(0, 90*24*3600)) 
                          for _ in range(num_rows)]

# Sắp xếp theo thời gian
df_pandas = df_pandas.sort_values('timestamp').reset_index(drop=True)

# 4. Xuất file
os.makedirs('../data', exist_ok=True)
df_pandas.to_csv('../data/web_logs.csv', index=False)
print(f"Đã tạo thành công 1 triệu dòng log tại ../data/web_logs_large.csv")
```

# Tạo dữ liệu bài 3

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os

# 1. Cấu hình
num_rows = 1000000
num_sensors = 100 # 100 cảm biến khác nhau

# 2. Sinh dữ liệu ngẫu nhiên
data = {
    'sensor_id': [f"S{np.random.randint(1, num_sensors + 1):03d}" for _ in range(num_rows)],
    # Nhiệt độ bình thường từ 20-60, nhưng có một số máy bị quá nhiệt
    'temperature': np.random.normal(40, 15, size=num_rows).round(1),
    # Độ ẩm 40-80%
    'humidity': np.random.uniform(40, 80, size=num_rows).round(1),
    # Độ rung bình thường < 30, thỉnh thoảng rung mạnh > 50
    'vibration': np.random.exponential(15, size=num_rows).round(2),
    'status': np.random.choice(['normal', 'warning', 'critical'], size=num_rows, p=[0.9, 0.07, 0.03])
}

df_pandas = pd.DataFrame(data)

# 3. Tạo timestamp trong 30 ngày gần đây
start_date = datetime(2026, 4, 1)
df_pandas['timestamp'] = [start_date + timedelta(seconds=np.random.randint(0, 30*24*3600)) 
                          for _ in range(num_rows)]

df_pandas = df_pandas.sort_values('timestamp').reset_index(drop=True)

# 4. Xuất file
os.makedirs('../data', exist_ok=True)
df_pandas.to_csv('../data/iot_sensors.csv', index=False)
print(f"Đã tạo thành công 1 triệu dòng dữ liệu IoT tại ../data/iot_sensors_large.csv")
```
