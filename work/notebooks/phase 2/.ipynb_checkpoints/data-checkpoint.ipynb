```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sqlalchemy import create_engine

print("Đang sinh dữ liệu mẫu (Phiên bản thực tế có phân hóa)...")
num_rows = 100000  # Nạp 100k dòng vào DB cho nhanh và nhẹ máy
num_stores = 50
num_products = 500
categories = ['Electronics', 'Grocery', 'Fashion', 'Home & Garden', 'Beauty']

# 1. Tạo sự phân hóa (Pareto) cho cửa hàng và sản phẩm
store_probs = np.random.power(0.5, size=num_stores)
store_probs = store_probs / store_probs.sum()

product_probs = np.random.power(0.3, size=num_products)
product_probs = product_probs / product_probs.sum()

# 2. Tạo danh mục và giá chuẩn cho sản phẩm
product_data = []
for i in range(1, num_products + 1):
    category = np.random.choice(categories, p=[0.1, 0.3, 0.3, 0.15, 0.15]) 
    if category in ['Electronics', 'Fashion']:
        price = round(np.random.gamma(shape=5, scale=50), 2) + 10 
    else:
        price = round(np.random.gamma(shape=2, scale=15), 2) + 5 
    product_data.append({'product_id': i, 'category': category, 'base_price': price})

product_df = pd.DataFrame(product_data)

# 3. Sinh dữ liệu giao dịch
data = {
    'transaction_id': np.arange(1, num_rows + 1),
    'store_id': np.random.choice(np.arange(1, num_stores + 1), size=num_rows, p=store_probs),
    'product_id': np.random.choice(np.arange(1, num_products + 1), size=num_rows, p=product_probs),
    # Phân bố số lượng mua, có chèn một ít số âm và số 0 để làm rác
    'quantity': np.random.choice([-2, -1, 0, 1, 2, 3, 4, 5, 8], size=num_rows, p=[0.01, 0.01, 0.02, 0.3, 0.3, 0.2, 0.1, 0.04, 0.02])
}
df = pd.DataFrame(data)

# Ghép danh mục và giá vào bảng giao dịch
df = df.merge(product_df, on='product_id', how='left')
df.rename(columns={'base_price': 'price'}, inplace=True)

# --- TẠO DỮ LIỆU LỖI (ANOMALIES) ĐỂ PYSPARK LÀM SẠCH ---
# Cố tình tạo 1000 dòng có giá trị Price lỗi (Âm hoặc Null)
error_indices = np.random.choice(num_rows, size=1000, replace=False)
df.loc[error_indices[:500], 'price'] = -15.5  # Giá âm
df.loc[error_indices[500:], 'price'] = np.nan # Null

# Cố tình tạo 500 dòng mất Store ID (Null)
null_store_indices = np.random.choice(num_rows, size=500, replace=False)
df.loc[null_store_indices, 'store_id'] = np.nan

# 4. Thêm thời gian
start_date = datetime(2025, 1, 1)
df['transaction_date'] = [start_date + timedelta(seconds=int(np.random.randint(0, 31536000))) for _ in range(num_rows)]

# 5. Đẩy vào PostgreSQL
print("Đang kết nối và đẩy dữ liệu vào PostgreSQL...")
engine = create_engine('postgresql://admin:password123@postgres:5432/retail_db')
df.to_sql('sales', engine, if_exists='replace', index=False)

print("Nạp thành công 100,000 dòng dữ liệu thực tế vào bảng 'sales'!")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import boto3
from io import BytesIO

print("1. Đang kết nối tới MinIO...")
s3_client = boto3.client('s3',
                         endpoint_url='http://minio:9000',
                         aws_access_key_id='admin',
                         aws_secret_access_key='password123')

bucket_name = 'logs'
try:
    s3_client.create_bucket(Bucket=bucket_name)
    print(f"-> Đã tạo bucket '{bucket_name}'.")
except Exception as e:
    print(f"-> Bucket '{bucket_name}' đã tồn tại.")

print("\n2. Đang sinh 1.000.000 dòng Web Log (Phân bố thực tế)...")
# --- CẤU HÌNH CỦA BẠN ---
num_rows = 1000000
num_users = 50000

pages = ['home', 'product_list', 'product_detail', 'cart', 'checkout', 'payment']
page_probs = [0.40, 0.25, 0.20, 0.08, 0.05, 0.02]

actions = ['view', 'scroll', 'click', 'add_to_cart']
action_probs = [0.50, 0.25, 0.20, 0.05]

devices = ['mobile', 'desktop', 'tablet']
device_probs = [0.70, 0.25, 0.05]

countries = ['Vietnam', 'USA', 'Thailand', 'Singapore', 'Japan', 'Korea']
country_probs = [0.60, 0.15, 0.10, 0.05, 0.05, 0.05]

user_probs = np.random.power(0.2, size=num_users)
user_probs = user_probs / user_probs.sum()
user_ids = [f"U{i:05d}" for i in range(1, num_users + 1)]

# --- SINH DỮ LIỆU ---
data = {
    'user_id': np.random.choice(user_ids, size=num_rows, p=user_probs),
    'page': np.random.choice(pages, size=num_rows, p=page_probs),
    'action': np.random.choice(actions, size=num_rows, p=action_probs),
    'device': np.random.choice(devices, size=num_rows, p=device_probs),
    'country': np.random.choice(countries, size=num_rows, p=country_probs)
}
df_pandas = pd.DataFrame(data)

# --- THÊM THỜI GIAN ---
start_date = datetime(2026, 1, 1)
df_pandas['timestamp'] = [start_date + timedelta(seconds=int(np.random.randint(0, 90*24*3600))) 
                          for _ in range(num_rows)]
df_pandas = df_pandas.sort_values('timestamp').reset_index(drop=True)

print("\n3. Đang nén và Upload thẳng lên MinIO (Sẽ mất khoảng 10-20 giây)...")
# Dùng BytesIO để giữ file trong RAM, không cần tạo file cứng trên ổ đĩa
csv_buffer = BytesIO()
df_pandas.to_csv(csv_buffer, index=False)
csv_buffer.seek(0)

# Tên file đẩy lên Data Lake
file_key = 'web_logs_large.csv'
s3_client.put_object(Bucket=bucket_name, Key=file_key, Body=csv_buffer.getvalue())

print(f"-> Upload XONG 1.000.000 dòng vào s3a://{bucket_name}/{file_key} !")
```
