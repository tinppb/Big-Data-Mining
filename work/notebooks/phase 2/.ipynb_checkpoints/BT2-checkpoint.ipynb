```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

print("1. Khởi tạo Spark với AWS S3/MinIO driver...")
# Dùng builder.config để nhồi cấu hình ngay lúc khởi tạo
spark = SparkSession.builder \
    .appName("MinIO_Log_Analysis") \
    .config("spark.jars.packages", "org.apache.hadoop:hadoop-aws:3.3.4,com.amazonaws:aws-java-sdk-bundle:1.12.262") \
    .config("spark.hadoop.fs.s3a.endpoint", "http://minio:9000") \
    .config("spark.hadoop.fs.s3a.access.key", "admin") \
    .config("spark.hadoop.fs.s3a.secret.key", "password123") \
    .config("spark.hadoop.fs.s3a.path.style.access", "true") \
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .config("spark.hadoop.fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider") \
    .getOrCreate()

print("\n2. Đang đọc file CSV trực tiếp từ MinIO Object Storage...")
# Đọc bằng giao thức s3a://
s3_path = "s3a://logs/web_logs_large.csv"
df_logs = spark.read.csv(s3_path, header=True, inferSchema=True)

# --- THỰC HIỆN CÁC YÊU CẦU PHÂN TÍCH ---
print(f"\n[A] Tổng số bản ghi log: {df_logs.count():,}")

print("\n[B] Phân tích số lượt truy cập theo Page:")
df_logs.groupBy("page").count().orderBy(F.desc("count")).show()

print("\n[C] Top 5 trang có số lượt truy cập cao nhất:")
df_logs.groupBy("page").count().orderBy(F.desc("count")).limit(5).show()

print("\n[D] Phân tích theo Thiết bị (Device):")
df_logs.groupBy("device").count().orderBy(F.desc("count")).show()

print("\n[E] Phân tích theo Quốc gia (Country):")
df_logs.groupBy("country").count().orderBy(F.desc("count")).show()

print("\n[F] Tính số lượt truy cập theo ngày...")
# Tạo cột date từ timestamp
df_logs = df_logs.withColumn("date", F.to_date(F.col("timestamp")))

# Aggregate theo ngày và chuyển về Pandas để lát vẽ biểu đồ
daily_traffic_pd = df_logs.groupBy("date") \
    .count().alias("daily_visits") \
    .orderBy("date") \
    .toPandas()
```
