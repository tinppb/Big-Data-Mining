# Phần A: Đọc dữ liệu

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Khởi tạo Spark
spark = SparkSession.builder.appName("IoTSensorAnalysis_Large").getOrCreate()

# --- Phần A. Đọc dữ liệu ---
df_iot = spark.read.csv("../data/iot_sensors.csv", header=True, inferSchema=True)

print("Schema dữ liệu cảm biến IoT:")
df_iot.printSchema()
print(f"Tổng số dòng dữ liệu: {df_iot.count()}")


```

# Phần B: Phân tích thống kê

```python

# 1. Thống kê nhiệt độ toàn hệ thống
print("\nThống kê nhiệt độ (Toàn nhà máy):")
df_iot.select(
    F.avg("temperature").alias("Nhiệt độ TB"),
    F.max("temperature").alias("Nhiệt độ Max"),
    F.min("temperature").alias("Nhiệt độ Min")
).show()

# 2. Độ rung trung bình theo từng cảm biến
print("Độ rung trung bình theo từng cảm biến (Top 10):")
df_iot.groupBy("sensor_id") \
    .agg(F.round(F.avg("vibration"), 2).alias("vibration_avg")) \
    .orderBy(F.desc("vibration_avg")) \
    .show(10)


```

# Phần C: Phát hiện bất thường

```python

# 1. Lọc các dòng có nhiệt độ > 80 hoặc độ rung > 50
anomalies = df_iot.filter((F.col("temperature") > 80) | (F.col("vibration") > 50))

print(f"\nSố lượng bản ghi bất thường phát hiện được: {anomalies.count()}")

# 2. Đếm số lượng bất thường theo sensor_id
print("Top 10 cảm biến có nhiều cảnh báo nhất:")
anomalies.groupBy("sensor_id").count().orderBy(F.desc("count")).show(10)


```

# Phần D: Phân tích thời gian

```python

# 1. Tạo cột ngày và tính nhiệt độ TB theo ngày
df_iot = df_iot.withColumn("date", F.to_date(F.col("timestamp")))

print("\nNhiệt độ trung bình theo ngày (Xu hướng):")
df_iot.groupBy("date") \
    .agg(F.round(F.avg("temperature"), 2).alias("daily_avg_temp")) \
    .orderBy("date") \
    .show()

# 2. Tính số lượng trạng thái status theo từng loại
print("Thống kê trạng thái vận hành:")
df_iot.groupBy("status").count().orderBy(F.desc("count")).show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
from pyspark.sql import functions as F


# Tạo cột date từ timestamp để phân tích theo ngày
df_iot = df_iot.withColumn("date", F.to_date(F.col("timestamp")))

# --- 1. SPARK AGGREGATION (Xử lý Big Data) ---

# A. Thống kê trạng thái toàn hệ thống
status_pd = df_iot.groupBy("status").count().toPandas()

# B. Lọc các bản ghi BẤT THƯỜNG (Temp > 80 HOẶC Vib > 50) và đếm theo Sensor
anomalies = df_iot.filter((F.col("temperature") > 80) | (F.col("vibration") > 50))
top_faulty_sensors_pd = anomalies.groupBy("sensor_id") \
    .count().alias("anomaly_count") \
    .orderBy(F.desc("count")) \
    .limit(10).toPandas()

# C. Tính nhiệt độ trung bình theo ngày
daily_temp_pd = df_iot.groupBy("date") \
    .agg(F.avg("temperature").alias("avg_temp")) \
    .orderBy("date").toPandas()

# D. Lấy chi tiết nhiệt độ của cái máy tệ nhất (để vẽ biểu đồ phân phối)
worst_sensor_id = top_faulty_sensors_pd['sensor_id'].iloc[0]
worst_sensor_data_pd = df_iot.filter(F.col("sensor_id") == worst_sensor_id) \
    .select("temperature", "vibration").toPandas()


# --- 2. VẼ BIỂU ĐỒ TRỰC QUAN CHUYÊN NGHIỆP ---
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Biểu đồ 1: Tỷ lệ Trạng thái (Donut Chart)
wedges, texts, autotexts = axes[0, 0].pie(
    status_pd['count'], labels=status_pd['status'], autopct='%1.1f%%', 
    colors=['#2ecc71', '#f1c40f', '#e74c3c'], startangle=90, pctdistance=0.85
)
# Vẽ vòng tròn trắng ở giữa để tạo Donut Chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
axes[0, 0].add_patch(centre_circle)
axes[0, 0].set_title('Tỷ Lệ Trạng Thái Hệ Thống', fontsize=14)

# Biểu đồ 2: Top 10 Cảm biến cảnh báo nhiều nhất (Bar Chart)
sns.barplot(ax=axes[0, 1], data=top_faulty_sensors_pd, x='sensor_id', y='count', palette='Reds_r')
axes[0, 1].set_title('Top 10 Cảm Biến Cần Bảo Trì Gấp', fontsize=14)
axes[0, 1].set_ylabel('Số lần vượt ngưỡng (>80°C hoặc >50 Hz)')
axes[0, 1].tick_params(axis='x', rotation=45)

# Biểu đồ 3: Xu hướng nhiệt độ trung bình theo ngày (Line Chart)
axes[1, 0].plot(daily_temp_pd['date'], daily_temp_pd['avg_temp'], marker='o', color='#3498db', linewidth=2)
axes[1, 0].set_title('Xu Hướng Nhiệt Độ Trung Bình Toàn Nhà Máy', fontsize=14)
axes[1, 0].set_ylabel('Nhiệt Độ (°C)')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, linestyle='--', alpha=0.6)

# Biểu đồ 4: Phân phối Mức độ rung của máy tệ nhất (Histogram/KDE)
sns.histplot(ax=axes[1, 1], data=worst_sensor_data_pd, x='vibration', bins=50, color='#e67e22', kde=True)
axes[1, 1].axvline(50, color='red', linestyle='dashed', linewidth=2, label='Ngưỡng nguy hiểm (>50)')
axes[1, 1].set_title(f'Phân Phối Độ Rung Của Máy Tệ Nhất ({worst_sensor_id})', fontsize=14)
axes[1, 1].set_xlabel('Độ rung (Hz)')
axes[1, 1].set_ylabel('Tần suất')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```
